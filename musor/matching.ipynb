```python
import pandas as pd
import numpy as np
pd.options.display.float_format = "{:.2f}".format
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_auc_score, f1_score
#from causalinference import CausalModel

original_data_filtered = pd.read_parquet('data/original_data_filtered.parquet')
original_data_filtered
```

```python
original_data_filtered = original_data_filtered.dropna()
original_data_filtered
```

```python
df_clear = original_data_filtered.select_dtypes(exclude=['object'])
df_clear = df_clear.drop(columns=['Регистрационный номер','Код налогоплательщика','index','Дата иска'])
df_clear.dtypes
```

```python
t = df_clear["Категориальная переменная"]
X = pd.get_dummies(df_clear.drop(columns=['Категориальная переменная']))
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic_classifier', LogisticRegression())
])#Задаем параметры конвеера
pipe.fit(X, t)#Обучаем модель (запускаем конвеер)
```

```python
threshold = 0.5
df_clear['proba'] = pipe.predict_proba(X)[:,1]#Записываем предсказание в столбец proba
df_clear['logit'] = df_clear['proba'].apply(lambda p: np.log(p/(1-p)) if p not in [0, 1] else np.nan)
df_clear['pred'] = np.where(df_clear['proba']>=threshold, 1, 0)#Создаем столбец Pred и по предсказанию записываем туда 1 или 0
df_clear['proba']
```

```python
Conflict ='Категориальная переменная'
print(f"Accuracy: {np.mean(df_clear[Conflict]==df_clear['pred']):.4f},\
 ROC AUC: {roc_auc_score(df_clear[Conflict], df_clear['proba']):.4f},\
 F1-score: {f1_score(df_clear[Conflict], df_clear['pred']):.4f}")#Считаем точность предсказания (в нашем исследовании не нужно)
# Visualise confusion matrix
pd.crosstab(df_clear[Conflict], df_clear['pred']).rename(columns={0: False,
                                                                  1:True})
```

```python
# Sort by 'logit' so it's quicker to find match
df_clear = df_clear
df = df_clear
df.sort_values('logit', inplace=True)
n = len(df)-1

for i, (ind, row) in enumerate(df.iterrows()):
    # Match the most similar untreated record to each treated record
    above = df.iloc[i:]
    control_above = above[~above[Conflict]]
    if not control_above.empty:

        match_above = control_above.iloc[0]
        distance_above = match_above['logit'] - row['logit']
        # Остальной код для сопоставления записей


        if row[Conflict]:
            # Find the closest untreated match among records sorted
            # higher. 'equal_or_above would' be more accurate but
            # used 'above' for brevity
            if i<n:
                above = df.iloc[i:]
                control_above = above[~above[Conflict]]

                match_above = control_above.iloc[0]
                distance_above = match_above['logit'] - row['logit']
                df.loc[ind, 'match'] = match_above.name
                df.loc[ind, 'distance'] = distance_above

            # Find the closest untreated match among records sorted
            # lower. 'equal_or_below' would be more accurate but
            # used 'below' for brevity
            if i>0:
                below = df.iloc[:i-1]
                control_below = below[~below[Conflict]]
                match_below = control_below.iloc[-1]
                distance_below = match_below['logit'] - row['logit']
                if i==n:
                    df.loc[ind, 'match'] = match_below.name
                    df.loc[ind, 'distance'] = distance_below

                # Only overwrite if match_below is closer than match_above
                elif distance_below<distance_above:
                    df.loc[ind, 'match'] = match_below.name
                    df.loc[ind, 'distance'] = distance_below
df[df[Conflict]]
```

```python
#Мэтчим выборки
indices = df[df['match'].notna()].index. \
    append(pd.Index(df.loc[df['match'].notna(), 'match']))
matched_df = df.loc[indices]
matched_df
```

```python
duplicates_count = matched_df.index.duplicated().sum()
print("Number of duplicates:", duplicates_count)

matched_df = matched_df[~matched_df.index.duplicated()]
```

```python
# Создайте DataFrame с сопоставленными записями

matched_df = pd.concat([original_data_filtered.loc[original_data_filtered.index.isin(indices)], matched_df], axis=1).reset_index(drop=True)

matched_df.columns
```

```python
# Транспонируйте DataFrame
matched_df_transposed = matched_df.T

# Удалите дубликаты
matched_df_transposed_no_duplicates = matched_df_transposed.drop_duplicates()

# Транспонируйте обратно
matched_df_no_duplicates = matched_df_transposed_no_duplicates.T
matched_df_no_duplicates
```

```python
matched_df_no_duplicates.to_excel('data/matched_data.xlsx', index=False)
matched_df_no_duplicates.to_parquet('data/matched_data.parquet', index=False)
```
